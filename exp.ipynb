```python
import pandas as pd
import numpy as np
import matplotlib.pyplot as plt
from SIV import SIV
```

```python
data = pd.read_csv('../StudentsPerformance.csv')
data.head()
```

```python
data = data[data['race/ethnicity'].isin(['group A', 'group B', 'group C'])]
data.drop(['reading score', 'writing score'], axis=1, inplace=True)
```

```python
# one hot encoding
data_encoded = pd.get_dummies(data, 
                              columns=['gender', 'race/ethnicity', 'parental level of education', 
                                       'lunch', 'test preparation course'],
                              drop_first=True)
data_encoded.head()
```

```python
data_encoded.shape
```

```python
siv = SIV()
X = data_encoded.drop('math score', axis=1)
Y = data_encoded['math score']
siv.fit(X, Y)
siv.get_coef()
```

```python
siv.get_X_hat()
```

```python
from sklearn.decomposition import FactorAnalysis
fa = FactorAnalysis(n_components=1)
fa.fit(X)
fa.components_
```
