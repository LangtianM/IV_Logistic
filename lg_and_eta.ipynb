```python
from IVClassification import *
import numpy as np
import matplotlib.pyplot as plt
import matplotlib

matplotlib.rcParams.update(matplotlib.rcParamsDefault)
```

# The Converged Value of Logistic Regression

## Beta and convergence

```python
beta_range = np.array([-2, -1, 0, 1, 2])
k = np.linspace(-1, 4, 100)
eta_range = np.power(10, k)

for beta in beta_range:
    
    lg_coef = np.array([])
    
    for eta in eta_range:
        data = sim_data(n=10000, beta = beta, eta=eta)
        
        lg = LogisticRegression()
        lg.fit(data.X.reshape(-1, 1), data.Y)
        lg_coef = np.append(lg_coef, lg.coef_)
        
    plt.plot(k, lg_coef, label = 'beta = ' + str(beta))
    plt.xticks(k[::10], [f'$10^{{{round(ki, 1)}}}$' for ki in k[::10]])

plt.xlabel('eta')
plt.ylabel('beta')
plt.legend()
```

## Gamma and Convergence

```python
gamma_range = np.array([-2, -1, 0, 1, 2])
k = np.linspace(-1, 4, 100)
eta_range = np.power(10, k)

for gamma in gamma_range:
    
    lg_coef = np.array([])
    
    for eta in eta_range:
        data = sim_data(n=10000, gamma = gamma, eta=eta)
        
        lg = LogisticRegression()
        lg.fit(data.X.reshape(-1, 1), data.Y)
        lg_coef = np.append(lg_coef, lg.coef_)
        
    plt.plot(k, lg_coef, label = 'gamma = ' + str(gamma))
    plt.xticks(k[::10], [f'$10^{{{round(ki, 1)}}}$' for ki in k[::10]])
    
plt.xlabel('eta')
plt.ylabel('beta')
plt.legend()
plt.axvline(x = 2, color = 'black', linestyle = '--')
```

```python
gamma_range = np.linspace(-8, 8, 100)
k = np.linspace(2, 4, 50)
eta_range = np.power(10, k)
stationary_coefs = np.array([])

for gamma in gamma_range:
    
    lg_coef = np.array([])
    
    for eta in eta_range:
        data = sim_data(n=10000, gamma = gamma, eta=eta)
        
        lg = LogisticRegression()
        lg.fit(data.X.reshape(-1, 1), data.Y)
        lg_coef = np.append(lg_coef, lg.coef_)
        
    stationary_coefs = np.append(stationary_coefs, np.mean(lg_coef))
    
    
plt.plot(gamma_range, stationary_coefs)
plt.xlabel('gamma')
plt.ylabel('converged beta')
```
